```python
import pandas as pd 
    
# making data frame from csv file 
data = pd.read_csv("Monthly_Food_Retail_Prices.csv") 
    
# making new data frame with dropped NA values 
new_data = data.dropna(axis = 0, how ='any') 
    
new_data
```

```python
new_data
```

```python
new_data[["Year"]] = new_data[["Year"]].apply(pd.to_numeric)
```

```python
new_data
```

```python
dataTypeSeries = new_data.dtypes
dataTypeSeries
```

```python
indexNames = new_data[ (new_data['Year'] >= 2001) & (new_data['Year'] <= 2015) ].index
new_data.drop(indexNames , inplace=True)
```

```python
new_data.to_csv('updated monthly food retail prices.csv')
```
